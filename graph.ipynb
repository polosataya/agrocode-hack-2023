```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Замените это вашим путем к файлу CSV
file_path = 'predict.csv'

# Загружаем данные из CSV файла
df = pd.read_csv(file_path)

# Создаем направленный граф
G = nx.DiGraph()

# Добавляем узлы и рёбра на основе данных в таблице
for i, row in df.iterrows():
    G.add_node(row['predict'])
    
    # Добавляем рёбра в соответствии с условиями
    if i > 0 and df.at[i-1, 'Резюме'] == row['Резюме']:
        G.add_edge(df.at[i-1, 'predict'], row['predict'])

# Визуализация графа
pos = nx.spring_layout(G)  # Вы можете выбрать другой алгоритм размещения
plt.figure(figsize=(15, 10))
nx.draw_networkx(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', font_size=8, edge_color='gray')

# Отображение графа в блокноте
plt.show()
```
